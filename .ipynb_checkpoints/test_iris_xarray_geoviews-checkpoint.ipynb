```python
ds1=xr.open_dataset(d1).load()
ssh1=ds1.isel(time=slice(100,125))
qmesh = gv.Dataset(ssh1.Grid_0001*0.01).to(gv.QuadMesh, ['NbLongitudes', 'NbLatitudes'], dynamic=True).redim.range(Grid_0001=(-1.0, 1.0),crs=crs.PlateCarree())
qmesh.opts(colorbar=True,cmap='jet')*gf.coastline()
ee = dataset.to(gv.Image, ['NbLongitudes', 'NbLatitudes'], 'Grid_0001')
```
